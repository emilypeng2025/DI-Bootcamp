Understanding Data Visualization

#Task: Explain why data visualization is important in data analysis.
As the name suggests, data visualization helps to make the data more easy to read for the whole process dealing with data: from data processing and pattern discovery, conclusion drawing to presenting the results. Whether there is an assumption or not, data visualization can uncover potential relationship or trends between variables. It will make the conclusion more presentable and easier to understand with charts and graphs.


#Task: Describe the purpose of a line graph in data visualization.
A line graph is used to display trends or changes in data over a continuous interval, often time. It connects individual data points with lines to show how a variable on the y-axis changes in relation to another variable on the x-axis. Line graphs are especially useful for identifying patterns, such as increases, decreases, or cycles over time.

ðŸŒŸ Exercise 4: Visualizing the Distribution of CGPA


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'CGPA': [3.0, 3.5, 2.8, 3.9, 3.7, 2.5, 3.3, 3.8, 3.4, 3.6, 3.6]
})

sns.histplot(data=df, x='CGPA', bins=5, color='skyblue', kde=True)
plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Number of Students')
plt.show()
```

ðŸŒŸ Exercise 5: Comparing Anxiety Levels Across Different Genders


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Do you have Anxiety?': ['No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
}
df = pd.DataFrame(data)

anxiety_df = df[df['Do you have Anxiety?'] == 'Yes']
sns.countplot(data=anxiety_df, x='Gender', palette='deep')
plt.title('Anxiety Levels Across Different Genders')
plt.xlabel('Gender')
plt.ylabel('The number of students with anxiety')
plt.show()
```
