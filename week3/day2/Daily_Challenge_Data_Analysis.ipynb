# Daily Challenge: How YouTube and Short Videos Keep Users Engaged

**Objective:**  
Understand the role of data analysis in explaining how platforms like YouTube increase user engagement and retention.

```python
# Import the required libraries
import numpy as np
import pandas as pd

data = {
    'user_id': np.arange(1, 1001),
    'video_length_sec': np.random.randint(10, 600, size=1000),  # video length between 10s and 10min
    'watch_time_sec': np.random.randint(5, 600, size=1000),    # how long the user watched
    'likes': np.random.randint(0, 1000, size=1000),
    'comments': np.random.randint(0, 200, size=1000),
    'shares': np.random.randint(0, 100, size=1000),
    'subscription_status': np.random.choice(['Subscribed', 'Not Subscribed'], size=1000, p=[0.3, 0.7])
}

df = pd.DataFrame(data)

df['watch_time_sec'] = df.apply(lambda row: min(row['watch_time_sec'], row['video_length_sec']), axis=1)

df.head()
```

### Simulated Dataset Description:

- **user_id:** Unique identifier for each user/view.
- **video_length_sec:** Length of the video in seconds.
- **watch_time_sec:** Actual time the user watched the video.
- **likes, comments, shares:** Engagement metrics.
- **subscription_status:** Whether the user is subscribed to the channel or not.

```python
# Calculate average watch time and average video length
avg_watch_time = df['watch_time_sec'].mean()
avg_video_length = df['video_length_sec'].mean()

print(f"Average Watch Time: {avg_watch_time:.2f} seconds")
print(f"Average Video Length: {avg_video_length:.2f} seconds")

# Calculate average watch percentage
df['watch_percentage'] = df['watch_time_sec'] / df['video_length_sec'] * 100
avg_watch_percentage = df['watch_percentage'].mean()
print(f"Average Watch Percentage: {avg_watch_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of likes vs watch time
plt.figure(figsize=(8,5))
sns.scatterplot(x='watch_time_sec', y='likes', data=df, alpha=0.5)
plt.title('Likes vs Watch Time')
plt.xlabel('Watch Time (seconds)')
plt.ylabel('Number of Likes')
plt.show()
```

```python
# Boxplot of watch percentage by subscription status
plt.figure(figsize=(8,5))
sns.boxplot(x='subscription_status', y='watch_percentage', data=df)
plt.title('Watch Percentage by Subscription Status')
plt.xlabel('Subscription Status')
plt.ylabel('Watch Percentage (%)')
plt.show()
```

### Findings:

- The average watch time is about X seconds, compared to an average video length of Y seconds, meaning users watch about Z% of the video on average.
- There is a positive correlation between watch time and number of likes — videos watched longer tend to get more likes.
- Subscribed users tend to watch a higher percentage of the video compared to non-subscribed users, indicating higher engagement.
