```python
import numpy as np

# Example survey data matrix (5 respondents, 4 questions)
survey_data = np.array([[5, 1, 4, 4],
                        [4, 2, 4, 3],
                        [5, 1, 1, 5],
                        [3, 2, 1, 2],
                        [2, 1, 2, 5]])

# Calculating various statistics
average_responses = np.mean(survey_data, axis=0)
total_responses = np.sum(survey_data, axis=0)
max_responses = np.max(survey_data, axis=0)
min_responses = np.min(survey_data, axis=0)

print("Average Responses:", average_responses)
print("Total Responses:", total_responses)
print("Maximum Responses:", max_responses)
print("Minimum Responses:", min_responses)
```

```python
import numpy as np
from scipy import stats

# Sample data (engagement time in minutes)
sample_data = temperature_data[:, 0]  # Example data

# Performing a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, popmean=20)
print('t_stat:', t_statistic)
print('p_value:', p_value)
# Interpreting the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm, expon, beta, t, uniform

# Set style
sns.set(style='whitegrid')

def describe_and_plot(data, title):
    print(f"{title}")
    print(f"Skewness: {skew(data):.2f}")
    print(f"Kurtosis: {kurtosis(data, fisher=True):.2f} (Fisher=False means normal == 3)")
    print("-" * 40)

    sns.histplot(data, kde=True)
    plt.title(title)
    plt.show()
```

```python
data_normal = np.random.normal(loc=0, scale=1, size=10000)
describe_and_plot(data_normal, "Normal Distribution")
```

```python
data_pos_skew = np.random.exponential(scale=1, size=10000)
describe_and_plot(data_pos_skew, "Positively Skewed (Exponential)")
```

```python
data_neg_skew = beta.rvs(a=2, b=10, size=10000)
describe_and_plot(data_neg_skew, "Negatively Skewed (Beta α=2, β=5)")
```

```python
data_leptokurtic = t.rvs(df=2, size=10000)
describe_and_plot(data_leptokurtic, "High Kurtosis (t-distribution, df=2)")
```

```python
data_platykurtic = np.random.uniform(low=-1, high=1, size=10000)
describe_and_plot(data_platykurtic, "Low Kurtosis (Uniform Distribution)")
```

```python
import numpy as np

# Generate daily dates for 2023
dates = np.arange(np.datetime64('2023-01-01'), np.datetime64('2024-01-01'))

# Use explicit np.datetime64 for comparison
jan_to_mar = dates[(dates >= np.datetime64('2023-01-01')) & (dates <= np.datetime64('2023-03-31'))]
apr_to_jun = dates[(dates >= np.datetime64('2023-04-01')) & (dates <= np.datetime64('2023-06-30'))]
jul_to_sep = dates[(dates >= np.datetime64('2023-07-01')) & (dates <= np.datetime64('2023-09-30'))]
oct_to_dec = dates[(dates >= np.datetime64('2023-10-01')) & (dates <= np.datetime64('2023-12-31'))]

# Example: print counts or first few entries
print("Jan–Mar:", jan_to_mar[:2], '...',jan_to_mar[-2:])
print("Apr–Jun:", apr_to_jun[:2], '...',apr_to_jun[-2:])
print("Jul–Sep:", jul_to_sep[:2], '...', jul_to_sep[-2:])
print("Oct–Dec:", oct_to_dec[:2], '...',  oct_to_dec[-2:])
```
